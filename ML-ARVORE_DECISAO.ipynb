# **ÁRVORE DE DECISÃO**

Uma árvore de decisão é um algoritmo de Machine Learning utilizado para resolver problemas de classificação divido um conjunto de dados original em dois ou mais subconjuntos de cada vez para melhor isolar as classes-alvo. Cada divisão pode ser representado graficamente como um nodo de árvore e traduzido em uma regra.

As sequências das divisões, ou seja, os ramos das árvores, definem um conjunto de regras que levam às classes-alvo isoladas.

image.png

A árvore de decisão consiste em três blocos distintos: nodos, ramificações, e folhas.

- Nodo : Identifica as características de divisão
- Ramificações : identificam valores diferentes das características de divisão
- Folhas : Associam a classe final a um caminho de decisão específico.

Como qualquer outro algoritmo de árvore de aprendizagem de máquina, as árvores de decisão são propensas ao overfitting.

Muitos critérios de divisão foram propostos, mas dois deles são implementados: O índice de Gini e o Gain Ratio. Árvores muito profundas podem levar a modelos que são muito detalhados e que não generalizam em dados novos. Por outro lado, árvores muito rasas podem levar a modelos muito simples que não ajustam os dados. O tamanho da árvore de decisão é crucial.

## **Gain Ratio**

O Gain Ratio é um conceito utilizado para medir informação ou desorganização e é  baseado na entropia, ou seja, medindo a puresa dos dados. A notação da entropia é seguida pela fórmula abaixo:

image.png

Onde:

- *p* : Vetor de probabilidades. Cada *pi* deve ser igual a 1, e todos os pi devem estar no intervalo [0,1];
- p ∈ Qn : Indica que o vetor p pertence ao conjunto de vetores de dimensão n com entradas racionais (Q), ou seja, os pi são geralmente números reais;
- log2pi : Logaritmo de base 2 de pi que mede a "surpresa" ou "informação" associada a observar o evento i. Os eventos mais improváveis carregam mais informação;
- pilog2pi : Pondera a quantidade de informação pelo quanto ela é provável. Ou seja, mede a contribuição média de cada evento para a incerteza total.

A entropia é a soma de todas as contribuições acima, com sinal negativo pois log2pi é negativo para pi E (0,1). Se todos os pi forem iguais ( Ex: Distribuição uniforme ), a entropia é máxima. Se um dos pi = 1 e os outros forem 0 ( ou seja, certeza sobre o resultado ), a entropia é mínima (zero).

image.png

O objetivo da árvore de decisão é isolar todos os subconjuntos puros com uma entropia de zero.Entretanto, na prática, uma árvore de decisão raramente encontra subconjuntos puros.
